```python
import pandas as pd

# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset_ML/diamonds.csv')

# Menampilkan dataframe
df.head()
```

#Box Plot

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom-kolom numerik untuk dibuat box plot
kolom_numerik = df.select_dtypes(include=['float64', 'int64'])

# Memilih fitur numerik
fitur_numerik = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat box plot untuk setiap fitur numerik
for fitur in fitur_numerik:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[fitur])
    plt.title(f'Box Plot of {fitur}')
    plt.show()
```

#Scatter Plot

```python
# Memilih fitur numerik
fitur_numerik = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat scatter plot untuk semua pasangan fitur numerik
sns.set(style="ticks")
sns.pairplot(df[fitur_numerik])
plt.show()
```

#Heat Map

```python
fitur_numerik = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menghitung korelasi antar fitur numerik
korelasi = df[fitur_numerik].corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Numeric Features')
plt.show()
```
